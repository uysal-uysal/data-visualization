```python
  import matplotlib.pyplot as plt
  import numpy as np 
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x),'-')
plt.plot(x, np.cos(x),'--')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend(('sin','cos'))
```

```python
fig = plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x),'-')
plt.xlabel('sin')

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x),'--', linewidth =4 , color='red')
plt.xlabel('cos')
```

```python
fig.savefig('data.png')
```

```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['petal_length'].plot()
```

```python
sns.lineplot(data = iris, y = 'petal_length', x = iris.index, hue='species')
```

```python
fig, ax = plt.subplots(figsize = (10,5))

ax.scatter(iris['sepal_length'], iris['sepal_width'])
ax.set_title('iris dataset')
ax.set_xlabel('sepal length')
ax.set_ylabel('sepal width')
```

```python
iris.groupby('species').mean()['sepal_length']
```

```python
data = iris.groupby('species').mean()['sepal_length']
ax = sns.barplot(x=data.index,y=data.values)
ax.set_title('sepal length means')
ax.set_ylabel('mean')
```

```python
sns.histplot(data=iris,x='sepal_width', bins=20)
```
